```python
finaldata = finaldata.sort_values(by=['match_id','innings'], ascending=[False,True])
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb1 = MultiLabelBinarizer()
df_encoded = pd.DataFrame(mlb1.fit_transform(finaldata['batters']), columns=mlb1.classes_, index=finaldata.index)
finaldata = pd.concat([finaldata, df_encoded], axis=1)
finaldata.drop('batters', axis=1, inplace=True)
Trained_classes0=mlb1.classes_
mlb2 = MultiLabelBinarizer()
df_encoded = pd.DataFrame(mlb2.fit_transform(finaldata['bowlers']), columns=mlb2.classes_, index=finaldata.index)
finaldata = pd.concat([finaldata, df_encoded], axis=1)
finaldata.drop('bowlers', axis=1, inplace=True)
Trained_classes1=mlb2.classes_





```
